```python
import types
def is_function(x):
    return type(x) == types.FunctionType

for x in dir(requests):
    if is_function(eval("requests.{}".format(x))):
        print(x)
```

```python
for x in dir(requests):
    y = eval("requests.{}".format(x))
    if not is_function(y) and not x.startswith('__'):
        print(x, type(y))
```
